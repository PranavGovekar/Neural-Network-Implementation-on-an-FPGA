```python
def add(x, y):
    maxlen = max(len(x), len(y))

    x = x.zfill(maxlen)
    y = y.zfill(maxlen)

    result = ''
    carry = 0

    for i in range(maxlen-1, -1, -1):
        r = carry
        r += 1 if x[i] == '1' else 0
        r += 1 if y[i] == '1' else 0

        result = ('1' if r % 2 == 1 else '0') + result
        carry = 0 if r < 2 else 1

    if carry != 0:
        result = '1' + result

    return result.zfill(maxlen)

def convert(binary):
    binary = '{:08b}'.format(binary)
    string = ""
    if(binary[0] == "-"):

        binary = binary.replace('-', '')

        binary = list(binary)

        for n in range(7):
            if(binary[n] == '0'):
                binary[n] = '1'
            else:
                binary[n] = '0'
        for num in binary:
            string = string+num

        out = add('1', string)
        
        out = "1" + out
        return out

    else:
        return binary


# string = '{:08b}'.format(67)
# twos = convert(-4)
# print(twos)

filename = "signedweights.txt"
file = open(filename, "w")

for array in l1_weights:
    for num in array:
        string = convert(num) + "\n"
        file.write(string)

file.close()
```
